# tfgraphviz

Examples

```python
import tensorflow as tf
import tfgraphviz as tfg
```

Building a tensorflow graph to execute addition.

```python
calc_g = tf.Graph()
```

```python
with calc_g.as_default():
    a = tf.constant(1, name="a")
    b = tf.constant(2, name="b")
    c = tf.add(a, b, name="add")
```

Visualize a graph with `tfg.board(...)`

```python
tfg.board(calc_g)
```

Building a tensorflow graph to execute linear regression.

```python
calc_reg = tf.Graph()
```

```python
with calc_reg.as_default():
    import numpy as np
    x_data = np.random.rand(100).astype(np.float32)
    y_data = x_data * 0.1 + 0.3
    W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
    b = tf.Variable(tf.zeros([1]))
    y = W * x_data + b
    loss = tf.reduce_mean(tf.square(y - y_data))
    optimizer = tf.train.GradientDescentOptimizer(0.5)
    train = optimizer.minimize(loss)
```

```python
tfg.board(calc_reg)
```

```python
tfg.board(calc_reg, depth=2)
```
